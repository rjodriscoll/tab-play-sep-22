```python
import pandas as pd
data = pd.read_csv('../data/train.csv')
```

```python
from plotnine import * # sorry 
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
data['year'] = data["date"].dt.year
data['month'] = data["date"].dt.month
data['day_of_week'] = data["date"].dt.day_of_week
data['is_month_end'] = data["date"].dt.is_month_end
data['is_year_end'] = data["date"].dt.is_year_end
```

```python


(
    ggplot(data, aes(x="factor(country)", y="num_sold"))
    + geom_boxplot(fill="green", alpha = 0.3, outlier_alpha = 0.2)
    + xlab('Country')
    + ylab('Number sold')
    + theme_light()  # change ticks labels on OX
)
```

```python
(
    ggplot(data, aes(y="num_sold", x= "factor(day_of_week)" ))
    + geom_boxplot(alpha = 0.3, outlier_alpha = 0.2,fill = 'green')
    + xlab('Day of week, monday = 0')
    + ylab('Number sold')
    + theme_light()  # change ticks labels on OX
)
```

```python
(
    ggplot(data, aes(x="factor(month)", y="num_sold"))
    + geom_boxplot(alpha = 0.3, outlier_alpha = 0.2, fill="green")
    + xlab('Month')
    + ylab('Number sold')
    + ggtitle('Monthly sales')
    + theme_light()  # change ticks labels on OX
    + theme(legend_title='Country')
)
```

```python
(
    ggplot(data, aes(x="factor(country)", y="num_sold"))
    + geom_boxplot(alpha = 0.3, outlier_alpha = 0.2, fill = 'green')
    + xlab('Country')
    + ylab('Number sold')
    + theme_light()  # change ticks labels on OX
)
```

```python
(
    ggplot(data, aes(x="factor(country)", y="num_sold", fill = 'factor(month)'))
    + geom_boxplot(alpha = 0.3, outlier_alpha = 0)
    + xlab('Country')
    + ylab('Number sold')
    + theme_light()  # change ticks labels on OX
)
```

```python
(
    ggplot(data, aes(x="factor(country)", y="num_sold", fill = 'factor(day_of_week)'))
    + geom_boxplot(alpha = 0.3, outlier_alpha = 0)
    + xlab('Country')
    + ylab('Number sold')
    + theme_light()  
    + labs(fill = 'Weekday (monday = 0)')
)
```

```python
(
    ggplot(data, aes(x="factor(country)", y="num_sold", fill = 'factor(is_year_end)'))
    + geom_boxplot(alpha = 0.3, outlier_alpha = 0)
    + xlab('Country')
    + ylab('Number sold')
    + theme_light() 
    + labs( fill='Is year end' )

)
```

```python
(
    ggplot(data, aes(x="factor(country)", y="num_sold", fill = 'factor(is_month_end)'))
    + geom_boxplot(alpha = 0.3, outlier_alpha = 0)
    + xlab('Country')
    + ylab('Number sold')
    + theme_light() 
    + labs( fill='Is month end' )

)
```
